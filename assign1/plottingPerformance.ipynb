### Question 5 - Plotting Parallel vs Serial Time

```python
import matplotlib.pyplot as plt
import pandas as pd

# Open results.csv
df = pd.read_csv('results.csv')

# Plot n versus Parallel Time and Serial Time
plt.plot(df['n'], df['Parallel Time'], label='Parallel Time')
plt.plot(df['n'], df['Serial Time'], label='Serial Time')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.title('n vs. Time')
plt.legend()
plt.show()
```
